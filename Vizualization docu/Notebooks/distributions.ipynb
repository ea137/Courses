```python
import matplotlib.animation as animation

def update(curr):
    if curr*100 == N:
        a.event_source.stop()

    title = ['Normal','Gamma','Exponential',
            'Uniform']
    XX = [x1,x2,x3,x4]
    colors = ['blue','orange','yellow','purple']
    for axee,xx,titre,color in zip(fig.axes, XX, title,colors):
        axee.cla()
        axee.hist(xx[:curr*100], density=True, bins=20, alpha=0.5,
                  color = color)
        axee.set_title(titre)
        axee.annotate(f'n = {curr*100}', [10,.5])
        

    plt.setp(fig.axes, xlim = (-7,21), ylim = (0,0.6))
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

N = 10000
# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(-2.5, 1, N)
x2 = np.random.gamma(2, 1.5, N)
x3 = np.random.exponential(2, N) + 7
x4 = np.random.uniform(-6,0, N)

# plot the histograms

fig , ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2)
fig.tight_layout(pad = 3)
plt.setp(fig.axes, xlim = (-7,21), ylim = (0,0.6))
a = animation.FuncAnimation(fig, update, interval=100)

#plt.text(x1.mean()-1.5, 0.5, 'x1\nNormal')
#plt.text(x2.mean()-1.5, 0.5, 'x2\nGamma')
#plt.text(x3.mean()-1.5, 0.5, 'x3\nExponential')
#plt.text(x4.mean()-1.5, 0.5, 'x4\nUniform')
None
```
